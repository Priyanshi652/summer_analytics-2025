```python
filter=df[(df["horsepower"] > 100) & (df["weight"] < 3000)]
common_origin=filter["origin"].value_counts().idxmax()
print("the most common origin for cars with horsepower > 100 and weight < 3000 :",common_origin)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median=df["mpg"].median()
filter=df[df["mpg"]>median]
filter["ratio"]=filter["horsepower"]/filter["weight"]
best_ratio=filter[filter["ratio"]==filter["ratio"].max()]
print(best_ratio[["name","ratio"]].to_string(index=False))
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns

average_mpg=df.groupby(["model_year","origin"])["mpg"].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=average_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

model_stats = df.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_std=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()

filtered = model_stats[(model_stats['appearances'] >= 2) & (model_stats['mpg_std'] < 1.0)]

min_std = filtered['mpg_std'].min()

lowest_variation_models = filtered[filtered['mpg_std'] == min_std]

result = lowest_variation_models.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

print(result[['name', 'appearances', 'avg_mpg',"mpg_std"]])
```
